```julia
DirViento8=collect(skipmissing(Datos8[:, 6])); 
DirViento9=collect(skipmissing(Datos9[:, 6])); 
DirViento10=collect(skipmissing(Datos10[:, 6])); 
DirViento11=collect(skipmissing(Datos11[:, 6])); 
DirViento12=collect(skipmissing(Datos12[:, 6])); 
DirViento13=collect(skipmissing(Datos13[:, 6])); 
DirViento14=collect(skipmissing(Datos14[:, 6])); 
DirViento15=collect(skipmissing(Datos15[:, 6])); 
DirViento16=collect(skipmissing(Datos16[:, 6])); 
DirViento17=collect(skipmissing(Datos17[:, 6])); 
DirViento18=collect(skipmissing(Datos18[:, 6])); 
DirViento19=collect(skipmissing(Datos19[:, 6]));
```

*** Enero ***

```julia
theta = DirViento8 #Dirección promedio del viento
r = VelViento8 #Vel prom del viento
width = 2pi/(length(theta))
fig = figure("1pyplot_windrose_barplot",figsize=(10,10)) # Crea una nueva figura
ax = axes(polar="true") # Crea una gráfica polar
title("Rosa de los vientos, Enero 2015")
b = bar(theta,r,width=width, color="darkred") # Gráfica de barras
dtheta = 20
ax[:set_thetagrids]([0:dtheta:360-dtheta;]) # Grid de 0 a 360 en incremento de theta
ax[:set_theta_zero_location]("N") # 0 grados en la parte superior
ax[:set_theta_direction](-1) # Sentido a las manecillas del reloj
fig[:canvas][:draw]() # Actualiza la figura
```
