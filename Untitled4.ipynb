*** Transformada de fourier***

```julia
vt=ones(200)*(1/200) 
cvt=filt(vt,1,Temp8); 
ftt = abs.(fft(cvt,1));
F8=(floor(length(ftt)/1 +1)/2)
fttp8 = ftt[2:convert(Int64,F8)]
fvec8 = collect(linspace(0,20,length(fttp8)))
pFt8 = plot(fvec8,fttp8, ylab="Amplitud", xlabel="Frecuencia (Hz)", box=:true, label="FFT de temperatura con filtro, Enero 2015", color="deepskyblue1", gridalpha=.4, grid=1.5, grid=:true, gridcolor="black")
```
