**DATOS DE ENERO**

```julia
#TEMPERATURA ENERO

#Cargar tema
theme(:ggplot2::Symbol) 

#plot Tiempo vs Temp
T20= plot(Tiempo20, Temp20, ylab="Temperatura (°C)", xlabel="Tiempo", title="Temperatura Enero", xrotation=40, leg=false) 

#Filtro media móvil
VT20=ones(200)*(1/200) #Ventana de tiempo
CVT20=filt(VT20,1,Temp20); #Convolución de datos y ventana de tiempo

#Plot con datos filtrados
TF20=plot(Tiempo20,CVT20, ylab="Temperatura (°C)",xlabel="Tiempo", title="Temperatura filtrada", xrotation=40, label=false, leg=false)

#Aplicando Fourier a los datos filtrados
ftt20 = abs.(fft(CVT20,1));
fttp20 = ftt20[2:convert(Int64, (length(ftt20)/2))] 
fvec20t = collect(linspace(0,20,length(fttp20)))
PF20= plot(fvec20t,fttp20, ylab="Amplitud", xlabel="Frecuencia (Hz)", title = " FFT Temperatura", label=false, leg=false)


#Espectograma
spe20=spectrogram(Temp20,2,1,fs=1)
fig20=heatmap(spe20.time,spe20.freq,spe20.power, color=:viridis, title="Espectrograma", xlabel="Tiempo", ylabel="Frecuencia (Hz)")

plot(T20, TF20, PF20, fig20)
```

```julia
#VELOCIDAD DEL VIENTO ENERO

#Cargar tema
theme(:ggplot2::Symbol) 

#plot Tiempo vs Vel viento
V20= plot(Tiempo20, VelViento20, ylab="Velocidad (m/s)", xlabel="Tiempo", title="Vel. Viento Enero", xrotation=40, leg=false) 

#Filtro media móvil
VV20=ones(200)*(1/200) #Ventana de tiempo
CVV20=filt(VV20,1,VelViento20); #Convolución de datos y ventana de tiempo

#Plot con datos filtrados
VF20=plot(Tiempo20,CVV20, ylab="Velocidad (m/s)",xlabel="Tiempo", title="Velocidad filtrada", xrotation=40, label=false, leg=false)

#Aplicando Fourier a los datos filtrados
ftt20V = abs.(fft(CVV20,1));
fttp20V = ftt20V[2:convert(Int64,(length(ftt20V)/2))] 
fvec20tV = collect(linspace(0,20,length(fttp20V)))
PFV20 = plot(fvec20tV,fttp20V, ylab="Amplitud", xlabel="Frecuencia (Hz)", title = " FFT Vel. Viento", label=false, leg=false)


#Espectograma
spe20V=spectrogram(VelViento20,2,1,fs=1)
fig20V=heatmap(spe20V.time,spe20V.freq,spe20V.power, color=:viridis, title="Espectrograma", xlabel="Tiempo", ylabel="Frecuencia (Hz)")

plot(V20, VF20, PFV20, fig20V)
```
