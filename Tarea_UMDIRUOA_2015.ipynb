```julia
L= plot(Tiempo8, Temp8 , label="Enero 2015", ylab="Temperatura(°C)", xlabel="Tiempo" , title="Temperatura " , box = true, label = "Enero 2015")
vt=ones(200)*(1/200)
cvt8=filt(vt,1,Temp8);
p8= plot(Tiempo8,cvt8 , label="Enero 2015", ylab="Temperatura(°C)", box=:true , xlabel="Tiempo" , title="Temperatura Filtrada" , label = "Enero 2015")
ftt = abs.(fft(cvt8,1));
F8=(floor(length(ftt)/1 +1)/2)
fttp8 = ftt[2:convert(Int64,F8)]
fvec8 = collect(linspace(0,20,length(fttp8)))
pFt8 = plot(fvec8,fttp8, ylab="Amplitud", xlabel="Frecuencia (Hz)", box=:true, gridalpha=.4, grid=1.5, grid=:true, title = " FFT Temperatura", label = "Enero 2015")
spe8=spectrogram(Temp8,2,1,fs=1)
fig8=heatmap(spe8.time,spe8.freq,spe8.power, color=:viridis, title="Espectrograma", xlabel="Tiempo", ylabel="Frecuencia (Hz)")

plot(L,p8,pFt8,fig8)
```

```julia

V8= plot(Tiempo8, VelViento8, ylab="Velocidad (m/s)", xlabel="Tiempo", title="Vel. Viento Enero", xrotation=40, leg=false) 
VV8=ones(200)*(1/200) #Ventana de tiempo
CVV8=filt(VV8,1,VelViento8); #Convolución de datos y ventana de tiempo
VF8=plot(Tiempo8,CVV8, ylab="Velocidad (m/s)",xlabel="Tiempo", title="Velocidad filtrada", xrotation=40, label=false, leg=false)
ftt8V = abs.(fft(CVV8,1));
FV8=(floor(length(ftt8V)/1 +1)/2) #Hacer el número par
fttp8V = ftt8V[2:convert(Int64,FV8)] 
fvec8tV = collect(linspace(0,20,length(fttp8V)))
PFV8 = plot(fvec8tV,fttp8V, ylab="Amplitud", xlabel="Frecuencia (Hz)", title = " FFT Vel. Viento", label=false, leg=false)
spe8V=spectrogram(VelViento8,2,1,fs=1)
fig8V=heatmap(spe8V.time,spe8V.freq,spe8V.power, color=:viridis, title="Espectrograma", xlabel="Tiempo", ylabel="Frecuencia (Hz)")

plot(V8, VF8, PFV8, fig8V)
```
